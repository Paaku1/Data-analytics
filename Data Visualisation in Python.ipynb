```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data=pd.read_csv('Netflix Userbase.csv')
data.head()
```

```python
data.describe()
```

```python
sns.countplot(x='Gender',data=data,hue='Gender',color='red')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Age',data=data,hue='Gender')
plt.title('Gender Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sub_counts=data['Subscription Type'].value_counts()
sub_counts
```

```python
sub_count=data['Subscription Type'].value_counts()
plt.pie(sub_count, labels=sub_count.index, autopct='%1.1f%%', startangle=90)
plt.title('Subscription Type')
plt.show()
```

```python
sub_count=data['Device'].value_counts()
plt.pie(sub_count, labels=sub_count.index, autopct='%1.1f%%', startangle=90)
plt.title('Subscription Type')
plt.show()
```

```python
sns.boxplot(x='Subscription Type',y='Monthly Revenue', data=data)
plt.title('Monthly Revenue by Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Monthly Revenue')
plt.show()
```

```python
x=data[['Monthly Revenue', 'Age']].corr()
sns.heatmap(x, annot=True, cmap='crest',)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.countplot(y='Country', data=data)
plt.title('Country Distribution')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Gender',y='Monthly Revenue', data=data)
plt.title('Monthly Revenue By gender')
plt.xlabel('Gender')
plt.ylabel('Montlyy Revenue')
plt.show()
```
